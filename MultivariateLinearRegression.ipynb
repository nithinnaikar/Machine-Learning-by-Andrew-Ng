```
# Importing Necessary Libraries
import numpy
import sklearn
import matplotlib
```

```
# Initializing the Dataset
dataset = numpy.loadtxt(fname = "/content/ex1data2.txt", delimiter = ",")
print("Dataset:\n\n{}\n\nShape:{}".format(dataset, dataset.shape))
```

```
# Dividing and Reshaping the Dataset Into Training and Testing Subsets
X_train, X_test, y_train, y_test = sklearn.model_selection.cross_val_scoretrain_test_split(dataset[:, :2], dataset[:, 2], test_size = 0.1)
X_train = numpy.resize(X_train, (42, 2))
y_train = numpy.resize(y_train, (42, 1))
X_test = numpy.resize(X_test, (5, 2))
y_test = numpy.resize(y_test, (5, 1))
```

```
# Visualizing the 42x2 Training Input Feaure Matrix 
print("X_train:\n\n{}\n\nShape:{}".format(X_train, X_train.shape))
```

```
# Visualizing the 42x1 Training Target Ouput Vector
print("y_train:\n\n{}\n\nShape:{}".format(y_train, y_train.shape))
```

```
# Visualizing the Training Data
figure = matplotlib.pyplot.figure()
graph = figure.add_subplot(projection = "3d")
graph.scatter3D(xs = X_train[:, 0], ys = X_train[:, 1], zs = y_train)
graph.set_xlabel("House Size")
graph.set_ylabel("Number of Bedrooms")
graph.set_zlabel("Price")
```

```
# Initializing the Model
model = sklearn.linear_model.LinearRegression(fit_intercept = True)
```

```
# Optimizing the Model
model.fit(X_train, y_train)
```

```
# Visualizing the Optimized Model
figure = matplotlib.pyplot.figure()
graph = figure.add_subplot(projection = "3d")
graph.scatter3D(xs = X_train[:, 0], ys = X_train[:, 1], zs = y_train)
xs = numpy.linspace(1000, 4500)
ys = numpy.linspace(1.0, 5.0)
zs = model.coef_[0][0] * xs + model.coef_[0][1] * ys + model.intercept_[0]
graph.plot3D(xs, ys, zs)
graph.set_xlabel("House Size")
graph.set_ylabel("Number of Bedrooms")
graph.set_zlabel("Price")
```
